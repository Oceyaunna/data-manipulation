## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!

import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np

baltimore_df = pd.read_csv("baltimore_9-24-2016_9-30-2017.csv")
boston_df = pd.read_csv("boston_9-24-2016_9-30-2017.csv")
new_york_df = pd.read_csv("new-york_9-24-2016_9-30-2017.csv")
philadelphia_df = pd.read_csv("philadelphia_9-24-2016_9-30-2017.csv")
```

```python
# Clean your data here!

# Repack Column
baltimore_df["Repack"] = baltimore_df["Repack"].replace({"N":False})
boston_df["Repack"] = boston_df["Repack"].replace({"N":False})
new_york_df["Repack"] = new_york_df["Repack"].replace({"N":False})
philadelphia_df["Repack"] = philadelphia_df["Repack"].replace({"N":False})

# Fill the Missing Values in the Type Column
baltimore_df["Type"] = baltimore_df["Type"].fillna("Conventional")
boston_df["Type"] = boston_df["Type"].fillna("Conventional")
new_york_df["Type"] = new_york_df["Type"].fillna("Conventional")
philadelphia_df["Type"] = philadelphia_df["Type"].fillna("Conventional")

# Drop the ones you decide are irrelevant using the code we used to drop the "Grade Column"
dropped_columns = ['Grade',
                   'Environment', 
                   'Quality', 
                   'Condition', 
                   'Appearance',
                   'Storage',
                   'Crop',
                   'Trans Mode',
                   'Origin District',
                   'Sub Variety',
                   'Item Size',
                   'Color',
                   'Commodity Name',
                   ]
baltimore_df = baltimore_df.drop(dropped_columns, axis=1)
boston_df = boston_df.drop(dropped_columns, axis=1)
new_york_df = new_york_df.drop(dropped_columns, axis=1)
philadelphia_df = philadelphia_df.drop(dropped_columns, axis=1)
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
combined_df = pd.concat([baltimore_df, boston_df, new_york_df, philadelphia_df])
combined_df.shape
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.

unit_of_sale_group = combined_df.groupby('Unit of Sale')
unit_of_sale_group.agg('mean')
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.

variety_group = combined_df.groupby(['City Name','Variety'])
variety_group.first()
variety_group.agg('mean')
#variety_group.agg({combined_df['Variety'].value_counts():'mean'})
```

```python
variety_counts = combined_df.value_counts('Variety')
avg_variety_counts = variety_counts/4

avg_variety_counts
```
